```python
df['polarity'] = df['transcript'].apply(lambda x : TextBlob(x).sentiment.polarity)
```

```python
df['transcript_len'] = df['transcript'].apply(lambda x: len(x))
df['word_count'] = df['transcript'].apply(lambda x:len(x.split()))
def get_avg_word_len(x):
  words = x.split()
  word_len=0
  for word in words:
    word_len = word_len + len(word)
  return word_len/len(words)
df['avg_word_len'] = df['transcript'].apply(lambda x:get_avg_word_len(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
x=  [df['transcript'][0]]
```

```python
vec = CountVectorizer().fit(x)
bow = vec.transform(x)
sum_words = bow.sum(axis=0) 
words_freq=[(word, sum_words[0,idx]) for word ,idx in  vec.vocabulary_.items()]
words_freq.sort(key = lambda x:x[1],reverse = True)
```

```python
def get_top_n_words(x,n,n_g=3):
    vec = CountVectorizer(ngram_range=(n_g,n_g),stop_words='english').fit(x)
    bow = vec.transform(x)
    sum_words = bow.sum(axis=0) 
    words_freq=[(word, sum_words[0,idx]) for word ,idx in  vec.vocabulary_.items()]
    words_freq.sort(key = lambda x:x[1],reverse = True)
    return words_freq[:n]
```

```python
words = get_top_n_words(list(df['transcript']),20,1)
df1 = pd.DataFrame(words,columns=['trigram','frequency'])
df1= df1.set_index('trigram')
df1.iplot(kind='bar',xTitle='unigram',yTitle='count', title='top 20 trigrams unigram')
```

```python
words = get_top_n_words(list(df['transcript']),20,2)
df1 = pd.DataFrame(words,columns=['trigram','frequency'])
df1= df1.set_index('trigram')
df1.iplot(kind='bar',xTitle='bigram',yTitle='count', title='top 20 trigrams bigram')
```

```python
words = get_top_n_words(list(df['transcript']),20,3)
df1 = pd.DataFrame(words,columns=['trigram','frequency'])
df1= df1.set_index('trigram')
df1.iplot(kind='bar',xTitle='trigram',yTitle='count', title='top 20 trigrams unigram trigram')
```
